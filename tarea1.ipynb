```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
muertes_totales= pd.read_csv('BaseCorregida.csv',delimiter=';')
```

```python
muertes_totales.dtypes
```

```python
muertes = muertes_totales
```

```python
muertes['FECHA_DEF'] = pd.to_datetime(muertes['FECHA_DEF'], dayfirst = True)
```

```python
muertes = muertes.sort_values(['FECHA_DEF'])
```

```python
muertes
```

```python
muertes=muertes.reset_index()
muertes=muertes.drop(['index'], axis=1 )
muertes
```

```python
grouper = pd.Grouper(key='FECHA_DEF', freq='W')
muertes["Semana"] = muertes.groupby(grouper).grouper.group_info[0] + 1
```

```python
muertes
```

```python
muertes_semana = muertes.groupby('Semana').size().reset_index(name='Muertes')
```

```python
muertes_semana
```

```python
plt.plot(muertes_semana['Semana'],muertes_semana['Muertes'])
plt.xlabel('Semana')
plt.ylabel('Muertes')
plt.title('Muertes totales por semana desde 2016')
plt.show()
```

```python
muertes_gripe = muertes[(muertes.GLOSA_CATEGORIA_DIAG1 == "Influenza [gripe] y neumonía")]
```

```python
muertes_gripe
```

```python
muertes_gripe=muertes_gripe.reset_index()
muertes_gripe
```

```python
muertes_gripe=muertes_gripe.drop(['index'], axis=1 )
```

```python
muertes_gripe_semana = muertes_gripe.groupby(['GLOSA_CATEGORIA_DIAG1','Semana']).size().reset_index(name='Muertes')
```

```python
muertes_gripe_semana
```

```python
plt.plot(muertes_gripe_semana['Semana'],muertes_gripe_semana['Muertes'],label='GRIPE')
plt.plot(muertes_semana['Semana'],muertes_semana['Muertes'])
plt.legend(loc="upper left") 
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos por Gripe por semana desde 2016')
plt.show()
```

```python
muertes_covid = muertes.drop(muertes[muertes.CODIGO_CATEGORIA_DIAG1 != "U00-U49"].index)
```

```python
muertes_covid
```

```python
muertes_covid = muertes_covid.reset_index()
muertes_covid = muertes_covid.drop(['index'], axis=1)
```

```python
muertes_covid_semana = muertes_covid.groupby(['Semana']).size().reset_index(name='Muertes')
```

```python
muertes_covid_semana
```

```python
plt.plot(muertes_covid_semana['Semana'],muertes_covid_semana['Muertes'],label='COVID19')
plt.plot(muertes_semana['Semana'],muertes_semana['Muertes'], label='TOTAL')
plt.legend(loc="upper left") 
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos por COVID19 por semana desde 2020')
plt.show()
```

```python
plt.plot(muertes_gripe_semana['Semana'],muertes_gripe_semana['Muertes'],label='GRIPE')
plt.plot(muertes_covid_semana['Semana'],muertes_covid_semana['Muertes'],label='COVID19')
plt.plot(muertes_semana['Semana'],muertes_semana['Muertes'], label='TOTAL')
plt.legend(loc="upper left") 
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos por COVID19 y GRIPE por semana')
plt.show()
```

Como se aprecia en el gráfico, las muertes de Covid semanalmente tienen un valor bastante mayor a las de la gripe común, por lo que no se puede aseverar que tengan el mismo impacto.
