```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos_riqueza_comuna = pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
datos_poblacion = pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/pobXedadCenso.csv")
```

Tras realizar la lectura de los archivos, nos damos cuenta que hay valores nulos en el primer archivo y por idosincrásia hay * en el segundo. Dado esto, se procederá a tratar estos valores para que no distorcionen el procesamiento general de los datos.

```python
datos_riqueza_comuna2 = datos_riqueza_comuna.replace(np.nan, datos_riqueza_comuna['idh'].mean())
```

```python
datos_riqueza_comuna2
```

```python
datos_poblacion2 = datos_poblacion.replace('*',0)
```

Ahora procederemos a seleccionar solamente los datos de nuestro interés. Es decir, las comunas correspondientes a la región Metropolitana

```python
datos_poblacion = datos_poblacion2['COMUNA'].isin(datos_riqueza_comuna['comuna'])
```

```python
datos_poblacion =datos_poblacion2.loc[datos_poblacion == True]
```

```python
datos_poblacion.dtypes
```

Como las edades están registradas como objetos, pasaremos a hacer la conversión a enteros

```python
def to_int(val):
    try:
        value = int(float(val))
    except ValueError:
        value = ""
    return value

datos_poblacion['EDAD_0A5'] = datos_poblacion['EDAD_0A5'].map(to_int)
datos_poblacion['EDAD_6A14'] = datos_poblacion['EDAD_6A14'].map(to_int)
datos_poblacion['EDAD_15A64'] = datos_poblacion['EDAD_15A64'].map(to_int)
datos_poblacion['EDAD_65YMAS'] = datos_poblacion['EDAD_65YMAS'].map(to_int)
```

```python
datos_poblacion = datos_poblacion.groupby('COMUNA').sum()
```

```python
datos_poblacion
```

```python
datos_poblacion['Total']=datos_poblacion.iloc[:,0:].sum(axis=1)
```

```python
datos_poblacion
```

Vemos que hay filas que poses solamente valores nulos. Procederemos a eliminarlos y a seleccionar solamente los correspondientes a las comunas de la región Metropolitana

```python
vacunacion_1ra = vacunacion_1ra[vacunacion_1ra['Codigo comuna'].notna()]
```

```python
comuna_1ra = vacunacion_1ra['Codigo comuna'].isin(datos_riqueza_comuna['comuna'])
```

```python
comuna_1ra =vacunacion_1ra.loc[comuna_1ra == True]
```

Juntaremos los totales vacunados según grupo etario

```python
comuna_1ra = comuna_1ra.drop(['Region','Codigo region', 'Comuna'],axis=1)
```

```python
comuna_total = comuna_1ra
```

```python
comuna_total['EDAD_15A64'] = comuna_total.iloc[:,2:52].sum(axis=1)
```

```python
comuna_total['EDAD_65YMAS'] = comuna_total.iloc[:,52:68].sum(axis=1)
```

```python
comuna_total
```

```python
comuna_total2 = comuna_total.drop(comuna_total.iloc[:,2:68], axis=1)
comuna_total2 = comuna_total2.rename(columns={'Codigo comuna':'COMUNA', 'EDAD_15A64':'Vacunados_15A64', 'EDAD_65YMAS':'Vacunados_65YMAS'})
```

```python
comuna_total2
```

```python
datos_poblacion.reset_index()
```

Como no hay registros de vacunados de gente menor a 15 años, los grupos etarios de EDAD_0A5 y EDAD_6A14 no serán considerados para el estudio.

```python
DATOS = datos_poblacion.drop(['EDAD_0A5','EDAD_6A14'], axis=1)
DATOS.reset_index()
```

```python
DATOS2 =pd.merge(DATOS, comuna_total2, on=['COMUNA'], how='left')
DATOS2
```

La columna 'Total' representa la población de la comuna según el censo, mientras que la columna 'Población' representa la cantidad de habitantes por comuna según los registros del MinCiencia. Como estos datos no coinciden, no es posible realizar una comparación entre los los habitantes y vacunados de los distintos grupos etarios de cada comuna. Por esto mismo, solamente se utilizará la columna 'Población' y se hará una relación porcentual entre los vacunados por grupo etario y el total de vacunados, además de la relación entre el total de vacunados y la población de la comuna

```python
DATOS2['Total_Vacunados'] = DATOS2['Vacunados_15A64']+DATOS2['Vacunados_65YMAS']
DATOS2['%Vacunados'] = DATOS2['Total_Vacunados']/DATOS2['Poblacion']*100
```

```python
DATOS2['HDI']=datos_riqueza_comuna2['HDI']
DATOS2
```

Ahora que ya se tienen todos la información necesaria recopilada en un solo dataframe, procederemos a hacer el estudio de correlación

```python
DATOS2.corr('pearson')
```

```python
import seaborn as sb

sb.lmplot(x = "HDI",
            y = "%Vacunados", 
            ci = None,
            data = DATOS2)
```

Respecto al gráfico anterior, el cual presenta el porcentaje de vacunados de las distintas comunas según su índice de desarrollo humano, se puede ver que existe una correlación positiva entre el índice de desarrollo humano y el porcentaje de la población que se vacuna, lo cual da una intuición para indicar que entre más rica sea una comuna, mayor será el porcentaje de gente que vive en ella que se vacunará, independiente de la edad.

```python
sb.lmplot(x = "HDI",
            y = "Vacunados_65YMAS", 
            ci = None,
            data = DATOS2)
```

En el gráfico recién mostrado, el cual representa la cantidad de vacunados de 65 años y más respecto a la riqueza de la comuna, al igual que el caso anterior, igual se aprecia una correlación positiva entre estas dos variables, lo cual permite indicar que entre mayor sea la riqueza de una comuna, mayor será la cantidad de gente de 65 años y más que se vacunará. Sin embargo, no todas las comunas tienen la misma cantidad de población de 65 años y más, por lo que se procederá a realizar la comparación entre la riqueza de la comuna y el porcentaje de vacunados de 65 años y más respecto al total de la población de la comuna.

```python
DATOS2['%Vacunados_15A64']=DATOS2['Vacunados_15A64']/DATOS2['Poblacion']*100
DATOS2['%Vacunados_65YMAS']=DATOS2['Vacunados_65YMAS']/DATOS2['Poblacion']*100
```

```python
sb.lmplot(x = "HDI",
            y = "%Vacunados_15A64", 
            ci = None,
            data = DATOS2)
sb.lmplot(x = "HDI",
            y = "%Vacunados_65YMAS", 
            ci = None,
            data = DATOS2)
```

Los dos gráficos anteriores muestran la relación entre el porcentaje de vacunados de los dos grupos etarios y el índice de desarrollo humano. Como se puede ver, existe una correlación positiva en ambos, pero en el caso de la población de 65 años y más, la pendiente de la recta es aúun mayor. Esto puede dar un indicio para aseverar que a mayor riqueza riqueza de la comuna, el aumento porcentual de la gente que se vacuna de 65 años y más es mayor que la de la gente entre 15 y 64 años. Una de las razones que podría justificar esto es que efectivamente a mayor índice de desarrollo humano tenga la comuna, hay más cantidad de gente viviendo en ellas que es de 65 años y más. Sin embargo, como los datos del censo no coinciden con los datos del MinCiencia, no podemos corroborarlo en su totalidad, pero con los datos que se tienen, se logra tener un indicio mayor para poder aseverar que el efecto de que las comunas más ricas se vacunaron primero son las que tienen mayor promedio de edad.
